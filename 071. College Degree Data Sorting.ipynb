```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors/page/"

page = 1
total_ranks_list = []
total_major_list = []
total_early_pay_list = []
total_mid_pay_list = []
total_high_value_list = []

while page <= 34:
    response = requests.get(f"{URL}1")
    webpage = response.text
    soup = BeautifulSoup(webpage, 'html.parser')

    ranks = [rank.getText() for rank in soup.find_all(name='td', class_='csr-col--rank')]
    majors = [major.getText() for major in soup.find_all(name='td', class_='csr-col--school-name')]
    career_pays = [career_pay.getText() for career_pay in soup.find_all(name='td', class_='csr-col--right')]

    rank_list = [rank.split(':')[1] for rank in ranks]
    major_list = [major.split(':')[1] for major in majors]

    early_pay_list = []
    mid_pay_list = []
    high_meaning_list = []

    for entry in career_pays:
        if entry.split(":")[0] == "Early Career Pay":
            early_pay_list.append(entry.split(":")[1])
        elif entry.split(":")[0] == "Mid-Career Pay":
            mid_pay_list.append(entry.split(":")[1])
        else:
            high_meaning_list.append(entry.split(":")[1])

    total_ranks_list.append(rank_list)
    total_major_list.append(major_list)
    total_early_pay_list.append(early_pay_list)
    total_mid_pay_list.append(mid_pay_list)
    total_high_value_list.append(high_meaning_list)

    page += 1

list_to_csv = []

for x in range(34):
    for y in range(24):
        data_to_csv = [total_ranks_list[x][y], total_major_list[x][y], total_early_pay_list[x][y],
                       total_mid_pay_list[x][y], total_high_value_list[x][y]]
        list_to_csv.append(data_to_csv)

df = pd.DataFrame(list_to_csv,
                  columns=['Rank', 'Major', 'Early Pay', 'Mid Pay', 'High Value'])
df.to_csv('highest_paying_jobs_updated.csv', index=False)
```
